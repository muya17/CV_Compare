```python
import random
import time

# Simulate receiving random commands
def generate_command():
    commands = ['M1x120', 'M2x80', 'S1x45', 'S2x90']
    return random.choice(commands)

while True:
    # Simulate message from PYNQ
    msg = generate_command()
    print(f"Received: {msg}")
    
    if msg.startswith("M1x"):
        speed = int(msg[3:])
        print(f"Set MOTOR 1 PWM to {speed}")
    elif msg.startswith("M2x"):
        speed = int(msg[3:])
        print(f"Set MOTOR 2 PWM to {speed}")
    elif msg.startswith("S1x"):
        angle = int(msg[3:])
        print(f"Set SERVO 1 angle to {angle}")
    elif msg.startswith("S2x"):
        angle = int(msg[3:])
        print(f"Set SERVO 2 angle to {angle}")
    else:
        print("Unknown command!")
        
    time.sleep(1)  # Simulate real-time loop
```
